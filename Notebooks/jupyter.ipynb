````python
#@title Haz clic ‚ñ∂ para ejecutar este c√≥digo
import numpy as np
import plotly.express as px

x = np.linspace(0, 10, 500)
y = np.sin(x)

fig = px.line(
    x=x, y=y,
    labels={'x':'t', 'y':'sin(t)'},
    title='Onda seno interactiva')
fig.show()
````

````python
#@title Haz clic ‚ñ∂ para ejecutar este c√≥digo
import numpy as np
import matplotlib.pyplot as plt
n = 10_000
plt.hist(np.random.normal(size=n), bins=50, alpha=0.6, label='Normal')
plt.hist(np.random.uniform(-3,3, size=n), bins=50, alpha=0.6, label='Uniform')
plt.legend(); plt.show()
````

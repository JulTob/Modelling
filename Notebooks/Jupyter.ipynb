# Celda
Un cuaderno es una secuencia de celdas. Cada celda contiene texto explicativo o cdigo ejecutable junto con su salida. Al hacer clic en una celda, esta pasa a ser el foco de trabajo y edicin.

## Celdas de cdigo

Debajo tienes una **celda de cdigo**. Una vez que el botn de la barra superior indique **CONNECTED**, haz clic dentro de la celda para seleccionarla y ejecutar su contenido de varias formas:

* Haz clic en el **icono de Play** del margen izquierdo de la celda.
* Pulsa **Cmd/Ctrl + Enter** para ejecutar la celda sin cambiar el foco.
* Pulsa **Shift + Enter** para ejecutar la celda y pasar a la siguiente (crendola si no existe).
* Pulsa **Alt + Enter** para ejecutar la celda e insertar una nueva celda de cdigo justo debajo.

Existen opciones adicionales para ejecutar algunas o todas las celdas desde el men **Runtime**.

## Narrativa computacional

````python
#@title Haz clic  para ejecutar este cdigo
import numpy as np
import plotly.express as px

x = np.linspace(0, 10, 500)
y = np.sin(x)

fig = px.line(
    x=x, y=y,
    labels={'x':'t', 'y':'sin(t)'},
    title='Onda seno interactiva')
fig.show()
````

````python
#@title Haz clic  para ejecutar este cdigo
import numpy as np
import matplotlib.pyplot as plt
n = 10_000
plt.hist(np.random.normal(size=n), bins=50, alpha=0.6, label='Normal')
plt.hist(np.random.uniform(-3,3, size=n), bins=50, alpha=0.6, label='Uniform')
plt.legend(); plt.show()
````

# Widgets y formularios

````python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

@widgets.interact(freq=(1,10,0.5))
def seno(freq=3):
    t = np.linspace(0, 2*np.pi, 400)
    plt.plot(t, np.sin(freq*t))
    plt.title(f'sin({freq}t)')
    plt.show()
````

## Magias

## Realizar Prueba de Hipótesis (Parte b)

### Subtask:
Determinar si el porcentaje de los que están 'A favor' es significativamente más pequeño que 50% con un nivel de confianza del 95%.


### Reasoning for Hypothesis Test (Part b):
To determine if the percentage of those 'A favor' is significantly smaller than 50% with a 95% confidence level, we will perform a one-tailed hypothesis test for proportions.

1.  **Formulate Hypotheses**:
    *   **Hypothesis nula (H0)**: La proporción de los que están a favor es igual o mayor al 50% (p >= 0.5).
    *   **Hypothesis alternativa (Ha)**: La proporción de los que están a favor es menor al 50% (p < 0.5).

2.  **Nivel de Significancia (alpha)**:
    *   Para un nivel de confianza del 95%, el nivel de significancia para una prueba unilateral es `alpha = 0.05`.

3.  **Calcular el Error Estándar bajo H0 (SE_0)**:
    *   Cuando se prueba una hipótesis sobre una proporción, el error estándar se calcula usando la proporción hipotética (p0) de la hipótesis nula. Aquí, p0 = 0.5.
    *   Fórmula: `SE_0 = sqrt(p0 * (1 - p0) / n)`

4.  **Calcular el Estadístico de Prueba Z**:
    *   Fórmula: `Z = (p_hat - p0) / SE_0`

5.  **Determinar el Valor Crítico Z**:
    *   Para una prueba unilateral izquierda con `alpha = 0.05`, encontramos el valor Z tal que el área a su izquierda es 0.05. Este valor se obtiene usando `norm.ppf(alpha)`.

6.  **Conclusión**:
    *   Si el estadístico de prueba Z es menor que el valor crítico Z, rechazamos la hipótesis nula (H0). De lo contrario, no la rechazamos.

**Reasoning**:
Now, I will execute the calculations for the hypothesis test in a code block, following the steps outlined in the previous markdown. This includes calculating the standard error under the null hypothesis, the Z-statistic, the critical Z-value, and finally determining whether to reject the null hypothesis.



```python
import numpy as np
from scipy.stats import norm

# Variables from previous steps:
# N = 10000
# n = 200
# p_hat = 0.45

# 1. Hypotheses (already stated in markdown)
# H0: p >= 0.5
# Ha: p < 0.5

# 2. Nivel de significancia (alpha)
alpha_ht = 0.05  # For 95% confidence level, one-tailed test

# 3. Proporción hipotética bajo H0
p0 = 0.5

# 4. Calcular el error estándar bajo H0 (SE_0)
SE_0 = np.sqrt(p0 * (1 - p0) / n)

# 5. Calcular el estadístico de prueba Z
Z_statistic = (p_hat - p0) / SE_0

# 6. Determinar el valor crítico Z para una prueba unilateral izquierda
Z_critical_ht = norm.ppf(alpha_ht)

# 7. Comparar el estadístico de prueba Z con el valor crítico Z y concluir
print(f"Proporción muestral (p_hat): {p_hat}")
print(f"Proporción hipotética (p0): {p0}")
print(f"Nivel de significancia (alpha): {alpha_ht}")
print(f"Error Estándar bajo H0 (SE_0): {SE_0:.4f}")
print(f"Estadístico de Prueba Z: {Z_statistic:.4f}")
print(f"Valor Crítico Z para alpha={alpha_ht} (unilateral izquierda): {Z_critical_ht:.4f}")

if Z_statistic < Z_critical_ht:
    print("Conclusión: Se rechaza la hipótesis nula (H0). Hay evidencia de que la proporción de los que están a favor es significativamente menor al 50%.")
else:
    print("Conclusión: No se rechaza la hipótesis nula (H0). No hay evidencia suficiente para afirmar que la proporción de los que están a favor es significativamente menor al 50%.")
```

## Calcular Tamaño Muestral Requerido (Parte c)

### Subtask:
Determinar el tamaño de muestra necesario para no cometer un error mayor de ±2 puntos porcentuales con un nivel de confianza del 95%.


### Reasoning for Sample Size Calculation (Part c):
To determine the necessary sample size, we will follow these steps:

1.  **Define the desired confidence level (95%)** and calculate the corresponding critical Z-value. For a 95% confidence, `alpha` is 0.05, and we need the Z-value for `1 - alpha/2` (i.e., `norm.ppf(0.975)`).
2.  **Define the maximum allowable margin of error (E)**. The problem specifies \u00b12 percentage points, which translates to E = 0.02 in decimal form.
3.  **Establish an estimated proportion (p)**. Since no prior estimate is available, we will use p = 0.5. This value maximizes the term `p * (1 - p)`, leading to the largest and most conservative (safe) sample size.
4.  **Apply the formula for the sample size** of a proportion:
    $$ n = \frac{Z^2 \cdot p \cdot (1-p)}{E^2} $$
5.  **Round the result up** to the nearest whole number, as the sample size must be an integer, and rounding up ensures that the error does not exceed the specified margin.
6.  Print the calculated sample size and all important intermediate values (Z, E, p).

**Reasoning**:
Now, I will implement the sample size calculation as described in the previous markdown block. This involves defining the confidence level, margin of error, and estimated proportion, then using the `norm.ppf` function to find the critical Z-value, applying the sample size formula, and finally rounding up the result to get the required sample size.



```python
import numpy as np
from scipy.stats import norm
import math

# 1. Nivel de confianza deseado y cálculo del valor crítico Z
confidence_level_c = 0.95
alpha_c = 1 - confidence_level_c
Z_critical_c = norm.ppf(1 - alpha_c / 2)

# 2. Margen de error máximo permitido (E)
E = 0.02  # 2 puntos porcentuales en formato decimal

# 3. Proporción estimada (p)
p = 0.5  # Usar 0.5 para el tamaño de muestra más conservador

# 4. Aplicar la fórmula para el tamaño de la muestra
n_calculated = (Z_critical_c**2 * p * (1 - p)) / (E**2)

# 5. Redondear el resultado al entero superior más cercano
n_required = math.ceil(n_calculated)

# 6. Imprimir los resultados
print(f"Nivel de confianza: {confidence_level_c * 100}%")
print(f"Valor Crítico Z: {Z_critical_c:.4f}")
print(f"Margen de Error (E): {E}")
print(f"Proporción estimada (p): {p}")
print(f"Tamaño de muestra calculado (sin redondear): {n_calculated:.2f}")
print(f"Tamaño de muestra requerido: {n_required}")
```
